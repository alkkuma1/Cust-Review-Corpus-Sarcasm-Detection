# EDA

#### Removing Punctuation

```python
df_less_dims['parent_comment'] = df_less_dims['parent_comment'].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))
```

```python
df_less_dims = df_less_dims[df_less_dims['parent_comment'] != ""]
df_less_dims.shape
```

#### Lowercase all the word

```python
df_less_dims['parent_comment'] = df_less_dims['parent_comment'].apply(lambda x: x.lower())
df_less_dims
```

#### Remove Stop Words

```python
nltk.download('stopwords')
def remove_stopwords(text):
    stop_words = set(stopwords.words("english"))
    words = word_tokenize(text)
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return " ".join(filtered_words)
```

```python
df_less_dims['parent_comment'] = df_less_dims['parent_comment'].apply(remove_stopwords)
```

```python
df_less_dims
```

#### Remove numbers

```python
df_less_dims['parent_comment'] = df_less_dims['parent_comment'].apply(lambda x: re.sub(r'[^\d]','',x))
```
